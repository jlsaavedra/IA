##Functions

```python
def matrix_distancias(n):
    distancias = [[0 for j in range(n)] for i in range(n)]
    for i in range(n):
        for j in range(i + 1, n):
            distancias[i][j] = distancias[j][i] = random.randint(1, 100)
    return distancias

def calcular_distancia(cities, distances):
  total_distance = 0
  for i in range(len(cities)-1):
    current_city = cities[i]
    next_city = cities[i+1]
    distance = distances[current_city-1][next_city-1]
    total_distance += distance
  return total_distance

def generar_ruta(numero_inicial, longitud):  
  ruta = [numero_inicial]
  while len(ruta) < longitud:
    numero_aleatorio = random.randint(1, longitud)
    if numero_aleatorio not in ruta:
      ruta.append(numero_aleatorio)
  return ruta

def generar_pob_rutas(numero_inicial, longitud):
  resultado = []
  while len(resultado) < longitud:
    temp = generar_ruta(numero_inicial, longitud)
    if temp not in resultado:
      resultado.append(temp)
  return resultado

def generar_ruta_bin(longitud, bin_code):
  numeros = []
  while len(numeros) < longitud * bin_code:
    numero_aleatorio = random.randint(0, 1)
    numeros.append(numero_aleatorio)
  return numeros

def generar_pob_bin(longitud, bin_code, cantidad):
  resultado = []
  while len(resultado) < cantidad:
    temp = generar_ruta_bin(longitud, bin_code)
    if temp not in resultado:
      resultado.append(temp)
  return resultado

def cotejar_bin2(ruta_actual, mi_array):
  pos_actual = 0  
  ruta_nueva = ruta_actual[:]

  for i in range(1, len(ruta_actual)):
    if mi_array[(i - 1) * 2] != mi_array[((i - 1) * 2) + 1]:
      if mi_array[(i - 1) * 2] == 0 and i + 1 < len(ruta_actual):
        temp = ruta_nueva[i]
        ruta_nueva[i] = ruta_nueva[i + 1]
        ruta_nueva[i + 1] = temp
  
  return ruta_nueva

def cotejar_bin3(ruta_actual, mi_array):
  pos_actual = 0  
  ruta_nueva = ruta_actual[:]

  for i in range(1, len(ruta_actual)):
    if mi_array[(i - 1) * 2] != mi_array[((i - 1) * 2) + 1]:
      if mi_array[(i - 1) * 2] == 0 and i + 1 < len(ruta_actual):
        temp = ruta_nueva[i]
        ruta_nueva[i] = ruta_nueva[i + 1]
        ruta_nueva[i + 1] = temp
      if mi_array[((i - 1) * 2) + 1] != mi_array[((i - 1) * 2) + 2]:
        if mi_array[((i - 1) * 2) + 1] == 0 and i + 2 < len(ruta_actual):
          temp = ruta_nueva[i]
          ruta_nueva[i] = ruta_nueva[i + 2]
          ruta_nueva[i + 2] = temp
  
  return ruta_nueva

def cruzar(parent1, parent2):
    idx1 = random.choice(range(1, len(parent1) - 1))
    idx2 = random.choice(range(1, len(parent2) - 1))
    child = [-1 for i in range(len(parent1))]
    child[idx1:idx2] = parent1[idx1:idx2]
    j = 0
    for i in range(len(child)):
        if child[i] == -1:
            while parent2[j] in child:
                j += 1
            child[i] = parent2[j]
    return child

def mutar(child):
    idx1 = random.choice(range(1, len(child) - 1))
    idx2 = random.choice(range(1, len(child) - 1))
    child[idx1], child[idx2] = child[idx2], child[idx1]
    return child

def generar_pob_cruzar(parent1, parent2, longitud, num_elite, mutation_prob):
  resultado = [parent1, parent2]
  while len(resultado) < longitud:
    temp_c = cruzar(parent1, parent2)
    if random.random() < mutation_prob:
      temp_c = mutar(temp_c)
    if temp_c not in resultado:
      resultado.append(temp_c)
  return resultado
```
